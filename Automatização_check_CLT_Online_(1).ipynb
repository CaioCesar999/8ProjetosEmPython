```python
df = pd.read_csv("Dezembro BI.csv")
df = df.drop(columns = ['col','Col 2','Col 3'], axis=1)
df['Dispendio RF'] = df['Dispendio RF'].apply(lambda x: x[3:]).astype(float)
df['Dispendio PEX'] = df['Dispendio PEX'].apply(lambda x: x[3:]).astype(float)
df = df.rename(columns={'Volume RF': 'Vol RF'})
df = df.rename(columns={'Volume PEX': 'Vol PEX'})
df['Vol PEX'] = df['Vol PEX'].astype(float)
df.head()
```

```python
df_teste = pd.read_excel("info_fonte.v7.xlsx", sheet_name = 'Dezembro')
```

```python
df_teste.head()
```

```python
df15 = df[['CLT','custo']]
for i in df.drop(columns=['CLT', 'custo']).columns:

  df[i] = pd.DataFrame(df[i] - df_teste[i])
  df15 = pd.concat([df15,df[i]],axis = 1)
  
df15
```

```python
# creating excel writer object
 
writer = pd.ExcelWriter('Resultado_Custo.xlsx') 
```

```python
# write dataframe to excel
 
df15.to_excel(writer, sheet_name='Dezembro Resultado')
 
# save the excel
writer.save()
```
